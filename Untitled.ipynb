split dataset

```python
from pathlib import Path
import random
```

```python
base_dir = 'data/blurred_sharp'

blur_images = list(map(lambda p: str(p.relative_to(base_dir)), Path(base_dir).glob('blurred/*.png')))
#sharp_images = list(map(to_relative, Path(base_dir).glob('sharp/*.png')))
random.shuffle(blur_images)
s = int(len(blur_images) * 0.8)
train_blur_images = blur_images[:s]
test_blur_images = blur_images[s:]
```

```python
# train_blur_imagesの名前からsharp画像へのパスを作る
train_sharp_images = [str(Path('sharp').joinpath(Path(f).name)) for f in train_blur_images]
test_sharp_images = [str(Path('sharp').joinpath(Path(f).name)) for f in test_blur_images]
```

```python
def save(n, x):
    with open(Path(base_dir).joinpath(n), 'w') as f:
        f.write('\n'.join(x))
```

```python
save('train_blur_images.txt', train_blur_images)
save('train_sharp_images.txt', train_sharp_images)
save('test_blur_images.txt', test_blur_images)
save('test_sharp_images.txt', test_sharp_images)
```
